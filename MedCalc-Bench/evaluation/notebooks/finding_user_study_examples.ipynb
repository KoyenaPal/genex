```python
import pandas as pd

def sort_by_column_length(csv_file, column_name, output_file=None):
    # Read the CSV file
    df = pd.read_csv(csv_file)

    # Ensure the column exists
    if column_name not in df.columns:
        raise ValueError(f"Column '{column_name}' not found in CSV.")

    # Sort by length of the column values
    df_sorted = df.iloc[df[column_name].astype(str).str.len().argsort()]

    # Save or return
    if output_file:
        df_sorted.to_csv(output_file, index=False)
    return df_sorted
```
