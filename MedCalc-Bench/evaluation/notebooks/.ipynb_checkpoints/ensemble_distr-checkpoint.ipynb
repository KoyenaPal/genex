```python
import os
import json
from collections import Counter
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.backends.backend_pdf import PdfPages
import seaborn as sns

# Set academic plotting style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams.update({
    'font.size': 12,
    'axes.titlesize': 14,
    'axes.labelsize': 12,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.titlesize': 16,
    'font.family': 'serif',
    'text.usetex': False,
})

# Model configuration - Colorblind friendly pastel colors
MODEL_NAMES = {0: 'QwQ', 1: 'OpenT', 2: 'DAPO', 3: 'GPT'}
MODEL_COLORS = {
    'QwQ': '#8ECAE6',      # Sky blue (colorblind safe)
    'OpenT': '#FFB3BA',    # Light pink (colorblind safe)
    'DAPO': '#FFDFBA',     # Light peach (colorblind safe)
    'GPT': '#BAFFC9'       # Light mint (colorblind safe)
}

def calculate_model_index_distribution(folder_path):
    """Calculate distribution of selected model indices from JSON files."""
    model_counts = Counter()
    total = 0
    
    if not os.path.exists(folder_path):
        print(f"Warning: Folder {folder_path} does not exist")
        return {}, Counter(), 0
    
    for filename in os.listdir(folder_path):
        if filename.endswith(".json"):
            filepath = os.path.join(folder_path, filename)
            
            try:
                with open(filepath, "r") as f:
                    data = json.load(f)
                
                if isinstance(data, list):
                    for row in data:
                        if isinstance(row, dict) and "selected_model_index" in row:
                            # Extract model name from the data and map to short form
                            selected_model = row["selected_model_index"]
                            if isinstance(selected_model, str):
                                # Map full model names to indices
                                if "Qwen/QwQ-32B" in selected_model or "QwQ" in selected_model:
                                    model_counts[0] += 1
                                elif "OpenThinker" in selected_model or "open-thoughts" in selected_model:
                                    model_counts[1] += 1
                                elif "DAPO" in selected_model or "BytedTsinghua" in selected_model:
                                    model_counts[2] += 1
                                elif "gpt-oss" in selected_model or "openai" in selected_model:
                                    model_counts[3] += 1  # Add GPT as index 3
                            else:
                                model_counts[selected_model] += 1
                            total += 1
                elif isinstance(data, dict) and "selected_model_index" in data:
                    selected_model = data["selected_model_index"]
                    if isinstance(selected_model, str):
                        if "Qwen/QwQ-32B" in selected_model or "QwQ" in selected_model:
                            model_counts[0] += 1
                        elif "OpenThinker" in selected_model or "open-thoughts" in selected_model:
                            model_counts[1] += 1
                        elif "DAPO" in selected_model or "BytedTsinghua" in selected_model:
                            model_counts[2] += 1
                        elif "gpt-oss" in selected_model or "openai" in selected_model:
                            model_counts[3] += 1
                    else:
                        model_counts[selected_model] += 1
                    total += 1
                    
            except (json.JSONDecodeError, KeyError) as e:
                print(f"Error processing {filename}: {e}")
                continue
    
    if total > 0:
        percentages = {model: (count / total) * 100 for model, count in model_counts.items()}
    else:
        percentages = {}
    
    return percentages, model_counts, total

def create_pie_charts(results_dict, output_pdf="model_distribution_analysis.pdf"):
    """Create publication-quality pie charts for all datasets."""
    
    with PdfPages(output_pdf) as pdf:
        n_datasets = len(results_dict)
        fig, axes = plt.subplots(1, n_datasets, figsize=(4*n_datasets, 4))
        
        if n_datasets == 1:
            axes = [axes]
        
        for idx, (dataset_name, results) in enumerate(results_dict.items()):
            ax = axes[idx]
            percentages = results['percentages']
            
            if not percentages:
                ax.text(0.5, 0.5, f'No data found', ha='center', va='center', 
                       transform=ax.transAxes, fontsize=12, style='italic')
                ax.set_title(dataset_name, fontsize=12, weight='bold')
                continue
            
            # Prepare data for pie chart
            sorted_items = sorted(percentages.items())
            labels = []
            colors = []
            sizes = []
            
            for model_idx, pct in sorted_items:
                model_name = MODEL_NAMES.get(model_idx, f'Model {model_idx}')
                labels.append(model_name)
                colors.append(MODEL_COLORS.get(model_name, '#D3D3D3'))
                sizes.append(pct)
            
            # Create pie chart with borders
            wedges, texts, autotexts = ax.pie(sizes, labels=labels, autopct='%1.1f%%', 
                                            startangle=90, colors=colors,
                                            wedgeprops={'edgecolor': 'black', 'linewidth': 0.5})
            
            # Style the text
            for autotext in autotexts:
                autotext.set_color('black')
                autotext.set_weight('bold')
                autotext.set_fontsize(9)
            
            for text in texts:
                text.set_fontsize(10)
                text.set_weight('bold')
            
            ax.set_title(dataset_name, fontsize=11, weight='bold', pad=10)
        
        plt.tight_layout(pad=2.0)
        pdf.savefig(fig, bbox_inches='tight', dpi=300)
        plt.show()

def print_summary(results_dict):
    """Print analysis summary."""
    print("\n" + "="*60)
    print("MODEL SELECTION DISTRIBUTION ANALYSIS")
    print("="*60)
    
    for dataset_name, results in results_dict.items():
        percentages = results['percentages']
        counts = results['counts']
        total = results['total']
        
        print(f"\nDataset: {dataset_name}")
        print("-" * 40)
        print(f"Total observations: {total:,}")
        
        if percentages:
            print("Model distribution:")
            for model_idx in sorted(percentages.keys()):
                model_name = MODEL_NAMES.get(model_idx, f'Model {model_idx}')
                pct = percentages[model_idx]
                count = counts[model_idx]
                print(f"  {model_name}: {count:,} observations ({pct:.2f}%)")
        else:
            print("No valid data found.")

if __name__ == "__main__":
    folders = {
        "QwQ + OpenT / GPT": "../medcalc_ensemble_outputs_gen_qwq_openthinker_eval_gpt_oss",
        "QwQ + GPT / DAPO": "../medcalc_ensemble_outputs_gen_qwq_gpt_oss_eval_dapo", 
        "QwQ + DAPO / GPT": "../medcalc_ensemble_outputs_gen_qwq_dapo_eval_gpt_oss"
    }
    
    # Process all datasets
    results_dict = {}
    for name, folder_path in folders.items():
        percentages, counts, total = calculate_model_index_distribution(folder_path)
        results_dict[name] = {
            'percentages': percentages,
            'counts': counts,
            'total': total
        }
    
    # Print summary and create visualizations
    print_summary(results_dict)
    create_pie_charts(results_dict)
    
    print(f"\n{'='*60}")
    print("ANALYSIS COMPLETE")
    print(f"{'='*60}")
```
