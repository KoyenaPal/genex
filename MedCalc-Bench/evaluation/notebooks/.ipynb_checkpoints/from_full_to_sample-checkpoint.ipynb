```python
import os
import json
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import rcParams
import numpy as np

# Set academic publication style
plt.style.use('seaborn-v0_8-whitegrid')
rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times New Roman', 'DejaVu Serif'],
    'font.size': 11,
    'axes.labelsize': 12,
    'axes.titlesize': 13,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'legend.fontsize': 10,
    'figure.titlesize': 14,
    'axes.linewidth': 0.8,
    'grid.linewidth': 0.5,
    'lines.linewidth': 1.5,
    'patch.linewidth': 0.8,
    'xtick.major.width': 0.8,
    'ytick.major.width': 0.8,
    'xtick.minor.width': 0.6,
    'ytick.minor.width': 0.6,
})

def get_short_name(model_full):
    """Convert model names to standardized short names for publication."""
    model_lower = model_full.lower()
    if "dapo" in model_lower:
        return "DAPO-32B"
    elif "qwq" in model_lower:
        return "QwQ-32B"
    elif "oss" in model_lower:
        return "OSS-20B"
    elif "nemotron" in model_lower:
        return "NRR-1.5B"
    elif "openthinker" in model_lower:
        return "OpenT-7B"
    else:
        return model_full

def extract_model_from_filename(filename):
    """Extract model name from filename."""
    # Remove file extension
    name = filename.replace('.json', '')
    
    # Split by underscores and look for model patterns
    parts = name.split('_')
    
    # Common model patterns in filenames
    model_patterns = ['dapo', 'qwq', 'oss', 'nemotron', 'openthinker']
    
    for part in parts:
        for pattern in model_patterns:
            if pattern in part.lower():
                return part
    
    # If no pattern found, try to extract from middle parts
    if len(parts) >= 3:
        return parts[1] + '-' + parts[2] if len(parts[1]) > 2 else parts[2]
    
    return filename  # fallback

# Data collection from both folders
data_list = []

# Define folders and their corresponding size labels
folders = {
    "../results": "Sample",
    "../results-original-full": "Full"
}

for folder_path, size_label in folders.items():
    if not os.path.exists(folder_path):
        print(f"Warning: Folder {folder_path} not found, skipping...")
        continue
        
    for fname in os.listdir(folder_path):
        if fname.endswith(".json"):
            fpath = os.path.join(folder_path, fname)
            
            try:
                with open(fpath, "r") as f:
                    data = json.load(f)
                    avg = data.get("overall", {}).get("average", None)
                    
                    if avg is not None:
                        # Determine condition based on filename
                        if "zero_shot_empty" in fname:
                            condition = "Empty CoT"
                        elif "zero_shot_original" in fname:
                            condition = "Default CoT"
                            
                            # For gpt-oss model, only include "low" variant
                            if "oss" in fname.lower():
                                if "low" not in fname.lower():
                                    continue  # Skip medium and high variants
                        else:
                            continue  # Skip files that don't match our patterns
                        
                        # Extract model name
                        model_name = extract_model_from_filename(fname)
                        
                        data_list.append({
                            "Model": model_name,
                            "Condition": condition,
                            "Size": size_label,
                            "Average": avg,
                            "Filename": fname  # Keep for debugging
                        })
                        
            except (json.JSONDecodeError, KeyError) as e:
                print(f"Error reading {fname}: {e}")
                continue

# Convert to DataFrame
df = pd.DataFrame(data_list)

if df.empty:
    print("No data found! Please check folder paths and filename patterns.")
    print("Expected patterns: 'zero_shot_empty' and 'zero_shot_original' in filenames")
    exit()

df["Model_short"] = df["Model"].apply(get_short_name)

# Print summary for verification
print("\nData Summary:")
print(df.groupby(['Model_short', 'Condition', 'Size'])['Average'].count())

# Define academic color palette - professional and colorblind-friendly
colors = {
    "Empty CoT": "#2166AC",     # Blue
    "Default CoT": "#5AAE61",   # Green
}

# Model ordering by parameter count
model_order = ["NRR-1.5B", "OpenT-7B", "OSS-20B", "QwQ-32B", "DAPO-32B"]
condition_order = ["Empty CoT", "Default CoT"]

# Filter to only include models we have data for
available_models = df["Model_short"].unique()
model_order = [m for m in model_order if m in available_models]

# Create the figure with proper academic proportions
fig = plt.figure(figsize=(12, 8))
gs = fig.add_gridspec(1, 2, width_ratios=[1, 1], hspace=0.05, wspace=0.3)

# Plot for each size condition
sizes = ["Sample", "Full"]
for i, size in enumerate(sizes):
    ax = fig.add_subplot(gs[0, i])
    
    # Filter data for this size
    size_data = df[df["Size"] == size]
    
    if size_data.empty:
        ax.text(0.5, 0.5, f'No data for {size}', 
                ha='center', va='center', transform=ax.transAxes)
        ax.set_title(f'{size} Dataset', fontweight='bold', pad=15)
        continue
    
    # Create horizontal bar plot
    y_positions = np.arange(len(model_order))
    bar_height = 0.35
    
    # Plot bars for each condition
    for j, condition in enumerate(condition_order):
        condition_data = size_data[size_data["Condition"] == condition]
        
        # Get values for each model (fill with NaN if missing)
        values = []
        for model in model_order:
            model_data = condition_data[condition_data["Model_short"] == model]
            if not model_data.empty:
                values.append(model_data["Average"].iloc[0])
            else:
                values.append(np.nan)
        
        # Plot bars
        bars = ax.barh(
            y_positions + j * bar_height - bar_height/2, 
            values, 
            bar_height, 
            label=condition,
            color=colors[condition],
            alpha=0.8,
            edgecolor='white',
            linewidth=0.5
        )
        
        # Add value labels on bars
        for k, (bar, value) in enumerate(zip(bars, values)):
            if not np.isnan(value) and value > 0:
                ax.text(
                    bar.get_width() + max(df["Average"]) * 0.01, 
                    bar.get_y() + bar.get_height()/2,
                    f'{value:.1f}',
                    ha='left', 
                    va='center',
                    fontsize=9,
                    fontweight='medium'
                )
    
    # Formatting
    ax.set_yticks(y_positions)
    ax.set_yticklabels(model_order)
    ax.set_xlabel('Overall Average Score', fontweight='medium')
    if i == 0:
        ax.set_ylabel('Model', fontweight='medium')
    else:
        ax.set_yticklabels([])
    
    # Set title with proper capitalization
    ax.set_title(f'{size} Dataset', fontweight='bold', pad=15)
    
    # Grid styling
    ax.grid(True, axis='x', alpha=0.3, linestyle='-', linewidth=0.5)
    ax.set_axisbelow(True)
    
    # Set x-axis limits for consistency
    max_val = df["Average"].max() if not df.empty else 100
    ax.set_xlim(0, max_val * 1.2)
    
    # Remove top and right spines for cleaner look
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(0.8)
    ax.spines['bottom'].set_linewidth(0.8)

# Add overall title
fig.suptitle('Model Performance Comparison Across Chain-of-Thought Conditions', 
             fontsize=14, fontweight='bold', y=0.95)

# Create legend
handles, labels = ax.get_legend_handles_labels()
legend = fig.legend(
    handles, labels, 
    loc='upper center', 
    bbox_to_anchor=(0.5, 0.02),
    ncol=2, 
    frameon=False,
    fontsize=10,
    title='Reasoning Condition',
    title_fontsize=11
)
legend.get_title().set_fontweight('bold')

# Adjust layout
plt.tight_layout(rect=[0, 0.08, 1, 0.92])

# Add figure caption
fig.text(0.1, 0.01, 
         'Figure 1: Performance comparison of language models under different chain-of-thought prompting conditions.\n' +
         'Sample data from ../results folder, Full data from ../results-original-full folder.',
         fontsize=9, style='italic', wrap=True, ha='left')

plt.show()

# Optional: Save as high-resolution figure for publication
# plt.savefig('model_comparison_academic.pdf', dpi=300, bbox_inches='tight', 
#             facecolor='white', edgecolor='none')
# plt.savefig('model_comparison_academic.png', dpi=300, bbox_inches='tight',
#             facecolor='white', edgecolor='none')
```
